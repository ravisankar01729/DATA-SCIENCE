```python
# Relational plots
# ---> Scatter plot
# ---> line plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# scatter plot
tips=sns.load_dataset('tips')
tips.head()
```

```python
# scatterplot->axes level -> graph as rectangular shape
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# relplot->figure level->graph as sqarish shape
# By using figure level functions we can replicate axes level function
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
# hue, style--> takes categorical data
# size--> takes numerical data
```

Facet plot

Q. On tips data set plot the scatterplot side by side for male & female

```python
tips
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',style='sex')
```

```python
# Facet plot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex') # if row='sex' then graphs will occupy top and bottom
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='smoker') # male smoker,male non-smoker,female smoker,female non-smoker
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='line',col='sex',row='smoker') # male smoker,male non-smoker,female smoker,female non-smoker
```

NOTE: Facet on work with figure level functions
