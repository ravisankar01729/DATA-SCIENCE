Categorical Plots--->Categorical Scatter plot, Categorical Distribution plot, Categorical Estimate Plot

Categorical Scatter plot---> Strip plot,Swarm plot.
Categorical Distribution plot----> Box plot, Violin plot.
Categorical Estimate plot--------> Bar plot, Point plot, Count plot.

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

```python
iris.head()
```

```python
# Strip plot---> plotted b/w catogerical data and numerical data
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False) # figure level function
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=0.5) # Seaborn adds additional noise by variable name jitter
```

```python
# fig level function---> catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',col='sex',hue='time')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm') # figure level function
```

```python
# axes level function
sns.swarmplot(data=tips,x='day',y='total_bill')
```

Categorical distribution plots(For 1 variable/column only), Univariate Analysis---->Box plot, Violin plot

```python
# Box plot
# iqr=q3(75%)-q1(25%)
# Calculated min=q1-1.5*iqr
# Calculated max=q3+1.5*1qr
# Potential Outliers lie outside of min and max
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(kind='box',data=tips,x='day',y='total_bill',hue='sex')
```

```python
# single boxplot--->numerical col
sns.boxplot(data=tips,y='total_bill')
```

```python
# Violin Plot(Box plot + kde plot)
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# Bar plot
sns.barplot(data=tips,x='sex',y='total_bill') # calculates average total_bill for males and females
```

```python
# Point plot=Barplot-bars
sns.pointplot(data=tips,x='sex',y='total_bill')
```

```python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python
# Count plot -> Special case of bar plot
sns.countplot(data=tips,x='sex')
```

```python
# har dhin kithney male aur female haya hei?
sns.countplot(data=tips,x='day',hue='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```
