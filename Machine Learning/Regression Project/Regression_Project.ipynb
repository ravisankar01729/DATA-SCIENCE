# Extract Independent features and Dependent features:

```python
df.columns
```

```python
cols=['Vehicle Class','Fuel Type','Transmission','Cylinders','Engine Size(L)','Fuel Consumption Comb (L/100 km)']
X=df[cols]
Y=df[['CO2 Emissions(g/km)']] # [[]]->to get as data frame, not as series.
```

```python
# Independent feature set
X.head(10)
```

```python
Y.head(10)
```

```python
print("Independent Feature Set shape :",X.shape)
print("Dependent Feature Shape       :",Y.shape)
```

```python
# Splitting the data into training(Seen data) and testing(Unseen data) sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2) # 30% of total for testing
```

```python
print('Traing Data shape     :',x_train.shape)
print('Training Labels shape :',y_train.shape)
print('Testing Data shape    :',x_test.shape)
print('Testing Labels shape  :',y_test.shape)
```

# Scaling (applied only on cont.. numerical columns)


```python
# 2 columns has different scales -> bring down data into similar scales.
# MinMax Scalar -> X_scaled=(X-X_min)/(X_max-X_min) -> Limits the data between 0 to 1.
cols = ['Engine Size(L)','Fuel Consumption Comb (L/100 km)']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Fitting on training set
scaler.fit(x_train[cols])
# Transform training and test sets
x_train[cols] = scaler.transform(x_train[cols])
x_test[cols]  = scaler.transform(x_test[cols])
```

# Encoding:

```python
cols=['Vehicle Class','Fuel Type',	'Transmission']
from sklearn.preprocessing import OrdinalEncoder # -> For multiple columns we use Ordinal Encoding,for 1 column we use label Encoding
encoder=OrdinalEncoder()
# Fitting on training set
encoder.fit(x_train[cols])
# Transform training and test sets
x_train[cols] = encoder.transform(x_train[cols])
x_test[cols]  = encoder.transform(x_test[cols])
```

```python
x_train.head(10)
```

```python
x_train['Fuel Type'].unique()
```

```python
x_train['Transmission'].unique()
```

# Training the machine learning model:

```python
from sklearn.neighbors import KNeighborsRegressor
regressor=KNeighborsRegressor(n_neighbors=3)
regressor.fit(x_train,y_train)
```

# Evaluating the model:

```python
pred_train = regressor.predict(x_train)
pred_test  = regressor.predict(x_test)

from sklearn.metrics import r2_score
print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score  : ', r2_score(y_test, pred_test))
```
