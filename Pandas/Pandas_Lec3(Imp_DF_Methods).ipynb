Q. Find the last match played by Virat Kohli in Delhi?

```python
ipl.head(3)
```

```python
mask=(ipl['Team1Players'].apply(lambda x: 'V Kohli' in x)) | (ipl['Team2Players'].apply(lambda x: 'V Kohli' in x))
kohli_played=ipl[mask]
kohli_played.head()
```

```python
kohli_played[kohli_played['City']=='Delhi'].sort_values('Date')
```

```python
kohli_played[kohli_played['City']=='Delhi'].sort_values('Date').drop_duplicates(subset=['City'],keep='last')
```
