# Questions on IPL dataset?

```python
ipl=pd.read_csv('/content/deliveries.csv')
ipl.head(3)
```

# Q. Find the top 10 batsman in terms of runs?

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

# Q. Find the batsman with maximum no of sixes?

```python
six_balls=ipl[ipl['batsman_runs']==6]
six_balls.head(3)
```

```python
six_balls.groupby('batsman')['match_id'].count().sort_values(ascending=False).head(1).index[0]
```

# Q. Find the batsman with most number of 4's and 6's in last 5 overs?

```python
last_fiveOvers=ipl[ipl['over']>15]
last_fiveOvers.head(3)
```

```python
last5_4_6=last_fiveOvers[(last_fiveOvers['batsman_runs']==4) | (last_fiveOvers['batsman_runs']==6)]
last5_4_6
```

```python
last5_4_6.groupby('batsman')['match_id'].count().sort_values(ascending=False).head(1).index[0]
```

Alternative Method:

```python
last5_4_6.value_counts('batsman').index[0]
```

# Q. Find V Kohli's record against all teams?

```python
kohli=ipl[ipl['batsman']=='V Kohli']
kohli
```

```python
kohli.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index() # reset_index() -> converts Series to DataFrame
```

# Q. Create a function that can return the highest score of any batsman?

```python
def high_score(batsman):
  temp_df=ipl[ipl['batsman']==batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
high_score('V Kohli')
```
