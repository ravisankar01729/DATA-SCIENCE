# Filtering a DataFrame

```python
# Find all the final winner in ipl data?
ipl.head(2)
```

```python
mask=ipl['MatchNumber']=='Final'
ipl[mask]
```

```python
ipl[mask][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured?
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# How many matches has CSK won in Kolkata?
ipl[(ipl['WinningTeam']=='Chennai Super Kings') & (ipl['City']=='Kolkata')].shape[0] # use bitwise not logical and
```

```python
# Toss winner is match winner in percentage ?
toss_matchWinner=ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]
total_matches=ipl.shape[0]
per=(toss_matchWinner/total_matches)*100
per
```

```python
# Movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)]['title_x']
```

```python
# Action movies with rating higher than 7.5?
mask1=movies['genres'].str.contains('Action')
mask2=movies['imdb_rating']>7.5
movies[mask1 & mask2]
```
