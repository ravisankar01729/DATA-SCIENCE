 # Q2. Find month by month revenue?

```python
temp_df=pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

# Q. Find top 3 stadiums with highest sixes/match ratio?

```python
matches.head()
```

```python
delivery.head()
```

```python
# DataFrame of deliveries with 6 runs
df=delivery[delivery['batsman_runs']==6]
df
```

```python
new_df=df.merge(matches,how='inner',left_on='match_id',right_on='id')
new_df
```

```python
num_sixes=new_df.groupby('venue')['venue'].count()
num_sixes
```

```python
num_matches=matches['venue'].value_counts()
num_matches
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

# Q. Find the Orange cap holder in all the seasons?

```python
data=matches.merge(delivery,how='inner',left_on='id',right_on='match_id')
data
```

```python
temp=data.groupby('season')
```

```python
for group1,data1 in temp:
  temp2=data1.groupby('batsman')
  print(group1,end=" :")
  dict={}
  for group2,data2 in temp2:
    dict[group2]=data2['batsman_runs'].sum()
  Keymax = max(dict, key= lambda x: dict[x])
  print(Keymax,"-->",dict[Keymax])
```
